```python
from sympy import symbols, cos, pi, summation, oo, plot, exp
from sympy.abc import x, n
```

```python
# Definimos la serie de Fourier en cosenos de f(x) = e^{2x}
a0 = (exp(2) - 1) / 2
an = (-2 * n * pi * ((-1)**n - 1)) / (4 + (n * pi)**2)
s = a0 + summation(an * cos(n * pi * x), (n, 1, oo))
```

```python
# Truncamos la serie para diferentes valores de n
s1 = a0 + sum(an.subs(n, k) * cos(k * pi * x) for k in range(1, 4))
s2 = a0 + sum(an.subs(n, k) * cos(k * pi * x) for k in range(1, 6))
s3 = a0 + sum(an.subs(n, k) * cos(k * pi * x) for k in range(1, 8))
```

```python
# Graficamos la serie y las aproximaciones
p = plot(s1, s2, s3, (x, -1, 1), show=False, legend=True)
p[0].label = 'n=3'
p[1].label = 'n=5'
p[2].label = 'n=7'
p.show()
```
