```python

!pip install networkx

# Importación de librerías necesarias
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Crear un grafo vacío
G = nx.Graph()

# Agregar nodos
G.add_node("A")
G.add_nodes_from(["B", "C", "D"])

# Agregar aristas (relaciones)
G.add_edge("A", "B")
G.add_edges_from([("A", "C"), ("B", "D"), ("C", "D")])

# Dibujar grafo
plt.figure(figsize=(6,4))
nx.draw(G, with_labels=True, node_color="skyblue", node_size=1000, font_size=14)
plt.show()
```

```python
G1 = nx.Graph()
G1.add_edges_from([(1,2),(2,3),(3,4)])
nx.draw(G1, with_labels=True, node_color="lightgreen", node_size=800)
plt.show()
```

```python
DG = nx.DiGraph()
DG.add_edges_from([(1,2),(2,3),(3,1)])
nx.draw(DG, with_labels=True, node_color="lightcoral", node_size=800, arrows=True)
plt.show()
```

```python
MG = nx.MultiGraph()
MG.add_edges_from([(1,2),(1,2),(2,3)])
nx.draw(MG, with_labels=True, node_color="lightyellow", node_size=800)
plt.show()
```

```python
print("Nodos:", G.nodes())
print("Aristas:", G.edges())
print("Número de nodos:", G.number_of_nodes())
print("Número de aristas:", G.number_of_edges())
```

```python
# Grado de cada nodo
print("Grados:", dict(G.degree()))

# Camino más corto
print("Camino más corto entre A y D:", nx.shortest_path(G, "A", "D"))

# Longitud del camino más corto
print("Longitud:", nx.shortest_path_length(G, "A", "D"))

# Centralidad de grado
print("Centralidad de grado:", nx.degree_centrality(G))
```

```python
plt.figure(figsize=(6,4))
pos = nx.spring_layout(G, seed=42)  # Layout de fuerza
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=1000, edge_color="gray", font_size=14)

# Destacar un nodo
nx.draw_networkx_nodes(G, pos, nodelist=["A"], node_color="orange", node_size=1200)
plt.show()
```

```python
# Creamos una red social ficticia
RS = nx.Graph()
RS.add_edges_from([
    ("Ana", "Luis"), ("Ana", "Marta"),
    ("Luis", "Carlos"), ("Marta", "Carlos"),
    ("Carlos", "Sofía"), ("Sofía", "Pedro"),
    ("Pedro", "Ana")
])

plt.figure(figsize=(7,5))
nx.draw(RS, with_labels=True, node_color="lightpink", node_size=1200, font_size=12, edge_color="purple")
plt.show()

print("Grado de conexión de cada persona:", dict(RS.degree()))
print("Camino más corto entre Ana y Sofía:", nx.shortest_path(RS, "Ana", "Sofía"))
```
