```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Definimos la condición inicial u(x,0) = x
def recta(x):
    y = np.zeros(len(x))
    for i in range(len(x)):
        y[i] = x[i]
    return y
```

```python
# Creamos el dominio espacial y temporal
x = np.linspace(0, 1, 1000)
t = np.linspace(0, 10, 1000)
f = recta(x)
```

```python
# Graficamos la condición inicial
plt.figure()
plt.plot(x, f, label='$u(x,0)=x$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Condición inicial (u(x,0)=f(x))")
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.01])
plt.grid()
plt.legend()
plt.show()
```

```python
# Definimos la serie de Fourier truncada para aproximar la condición inicial
def u_0(x, N=20):  # N es el número de armónicos
    f = np.zeros(len(x))
    for n in range(1, N):
        f += (-1) ** (n + 1) * np.sin(n * np.pi * x) / n  # Sumamos los términos de la serie
    return (2.0 / np.pi) * f
```

```python
# Calculamos la aproximación con N=20 términos
f_n = u_0(x)
```

```python
# Graficamos la aproximación junto con la condición inicial
plt.figure()
plt.plot(x, f_n, label='$f_N(x)$ con $N=20$')
plt.plot(x, f, label='$f(x)=x$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Aproximación de la condición inicial")
plt.xlim([-0.01, 1.1])
plt.ylim([-0.01, 1.2])
plt.grid()
plt.legend()
plt.show()
```

```python
# Visualizamos la aproximación para diferentes cantidades de armónicos
plt.figure()
for i in range(1, 20, 5):
    plt.plot(x, u_0(x, i), label='$f_N(x)$ con $N=%d$' % i)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Aproximación de la condición inicial")
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.2])
plt.grid()
plt.legend()
plt.show()
```

```python
# Definimos la solución completa u(x,t) agregando la evolución temporal
def u_x_t(x, t, N=50):
    f = np.zeros((len(x), len(t)))
    for i in range(len(t)):
        for n in range(1, N):
            f[:, i] += (-1) ** (n + 1) * np.sin(n * np.pi * x) / n * np.exp(-n ** 2 * np.pi ** 2 * t[i])
    return (2.0 / np.pi) * f
```

```python
# Evaluamos la solución en el tiempo
U = u_x_t(x, t)
```

```python
# Graficamos la solución en el instante inicial t=0
plt.figure()
plt.plot(x, U[:, 0], label='$u(x,0)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Aproximación de la condición inicial")
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.2])
plt.grid()
plt.legend()
plt.show()
```

```python
# Graficamos la evolución de la solución en distintos tiempos
plt.figure()
plt.plot(x, U[:, 0], label='$u(x,t=0)$')
plt.plot(x, U[:, 1], label='$u(x,t=1)$')
plt.plot(x, U[:, 2], label='$u(x,t=2)$')
plt.plot(x, U[:, 3], label='$u(x,t=3)$')
plt.plot(x, U[:, 10], label='$u(x,t=10)$')
plt.plot(x, U[:, 20], label='$u(x,t=20)$')
plt.xlabel('$x$')
plt.ylabel('$u(x,t)$')
plt.title("Evolución de la solución")
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.2])
plt.grid()
plt.legend()
plt.show()
```

```python
# Graficamos la solución en 3D para ver la evolución en el tiempo
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection="3d")
```

```python
# Creamos la malla espacial y temporal
T, X = np.meshgrid(t, x) 
X, T = np.meshgrid(x, t, indexing='ij')  # Se agrega indexing='ij' para que coincida con U
```

```python
# Graficamos la solución en 3D para ver la evolución en el tiempo
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection="3d")

# Graficar la superficie
ax.plot_surface(T, X, U, cmap="inferno", edgecolor='none')

plt.show()
```

```python
# Configuración de la gráfica en 3D
ax.set_xlabel('$t$')
ax.set_ylabel('$x$')
ax.set_zlabel('$u(x,t)$')
ax.set_title("Solución de la ecuación de calor")

# Graficar la superficie
ax.plot_surface(T, X, U, cmap="inferno", edgecolor='none')

plt.show()
```
